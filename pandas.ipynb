Series
一種一維數據結構，可以儲存任何數據類型，每個Series都會有一個數據類型來定義其值的類型。

```python
s = pd.Series([10, "20", True, [1, 2, 3], {"key": "value"}])
print(s)
```

DataFrame
一種二維數據結構，可以看作是由多個Series組成的表格，每列可以有不同的數據類型。

```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"],
}
df = pd.DataFrame(data)
print(df)
```

DataFrame創建方式：
1. 從字典（dict）列表創建
2. 從列的字典（dict）創建
3. 從行的列表創建
4. 從 Series 字典創建

```python
# 從字典（dict）列表創建
data1 = [
    {"Name": "David", "Age": 28, "City": "Miami"},
    {"Name": "Eva", "Age": 22, "City": "Seattle"},
    {"Name": "Frank", "Age": 33, "City": "Austin"},
]

df = pd.DataFrame(data1)
print(df)
```

```python
# 從列的字典（dict）創建
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [24, 30, 28],
    "Job": ["Engineer", "Designer", "Teacher"],
}

df = pd.DataFrame(data)
print(df)
```

```python
# 從行的列表創建
rows_data = [
    ['Alice', 24, 'Engineer'],
    ['Bob', 30, 'Designer'],
    ['Charlie', 28, 'Teacher']
]
col_names = ['Name', 'Age', 'Job']
df = pd.DataFrame(rows_data, col_names)

print(df)
```

```python
# 從 Series 字典創建
s1 = pd.Series([1, 2, 3])
s2 = pd.Series(["Alice", "Bob", "Charlie"])
s3 = pd.Series([24, 30, 28])

df = pd.DataFrame({"id": s1, "name": s2, "age": s3})
print(df)
```

```python
# 創建一個簡單的 DataFrame
data = {
    "Integers": [1, 2, 3, 4],
    "Floats": [0.1, 0.2, 0.3, 4],
    "Strings": ["apple", "banana", "cherry", "123"],
}
df = pd.DataFrame(data)

# 使用屬性
print("1 Data Types:\n", df.dtypes)
print()
print("2 Shape:", df.shape)
print()
print("3 Index:", df.index)
print()
print("4 Columns:", df.columns)
print()
print("5 Info:", df.info())
print()
print("6 Describe:", df.describe())
print()
print("7 Head:\n", df.head())
print()
print("8 Size:", df.size)
print()
print("9 Is Empty:", df.empty)
```

處理 DataFrame 資料

1. 直接賦值
df['new_column'] = [value1, value2, value3, ...]

```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"],
}
df = pd.DataFrame(data)
print(df)
print()

# df["Score"] = [88, 92, 79] # 1. 直接賦值
# df["Country"] = "USA" # 2. 單一值賦值
# df["Age"] = df["Age"] * 10 # 3. 基於現有列進行計算賦值
df["AgeGroup"] = pd.cut(
  df["Age"],
  bins=[0,29,39,120],
  labels=["20s-","30s","40+"]
)
print(df)
```

```python
ages = pd.Series([0, 18, 29, 30, 39, 40, 120, 121])

out = pd.cut(ages,
  bins=[0, 29, 39, 120],
  labels=["20s-", "30s", "40+"],
  include_lowest=True,  # 讓 0 被納入第一箱
  right=True
)

print(out)
# print(out.value_counts(dropna=False))
```

使用concat()合併

```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"],
}
df = pd.DataFrame(data)
print(df)
new_rows = pd.DataFrame([
    {"Name":"Evan","Age":33,"City":"Boston"},
    {"Name":"Fiona","Age":41,"City":"Austin"},
])

df = pd.concat([df, new_rows], ignore_index=True)
print()
print(df)
```

```python
data = {'apples': [3, 2], 'oranges': [4, 6]}
df = pd.DataFrame(data)
df['bananas'] = [5, 3]  # 增加一列 bananas

new_row = {'apples': 7, 'oranges': 8, 'bananas': 2}
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
print(df)
```

DataFrame 數據清理

去除重複數據

```python
data = {
    "Name": ["Alice","Bob","Charlie","Alice","Bob"],
    "Age":  [25,30,35,25,31],
    "City": ["NY","LA","CHI","NY","LA"]
}
df = pd.DataFrame(data)

print(df)

# df1 = df.drop_duplicates() # 移除完全重複的列
# print()
# print(df1)
# df2 = df.drop_duplicates(subset=["Name","City"], keep="first") # 根據 Name 和 City 欄位移除重複列，保留第一次出現的
# print()
# print(df2)
mask = df.duplicated(subset=["Name","City"], keep="first")
dupes = df[mask]          # 重複的那些列
unique = df[~mask]        # 去重後保留的那些列
print()
print(dupes)
print()
print(unique)
```

填充
用固定值、平均數、中位數或其他算法預測的值填充缺失值。

```python
import numpy as np

df = pd.DataFrame({
  'A': [1, np.nan, 3],
  'B': [4, 5, np.nan],
  'C': [np.nan, np.nan, np.nan]
})
print(df); print()

# 使用固定值填充（保持原樣）
df_filled = df.fillna(0)
print(df_filled); print()

# 使用各列（欄位）的平均值填充（保持原樣）
df_filled = df.fillna(df.mean(numeric_only=True))
print(df_filled); print()

# 向前填充 —— 舊：df.fillna(method='ffill')
df_filled = df.ffill()
print(df_filled); print()

# 若想改成向後填充
df_filled_b = df.bfill()
print(df_filled_b); print()
```
